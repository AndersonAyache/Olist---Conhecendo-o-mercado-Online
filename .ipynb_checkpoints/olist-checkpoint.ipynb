# Introdução

A modalidade de vendas online já se consolidou em diversos modelos de negócio e continua a crescer a cada dia. Este estudo tem como propósito analisar as vendas realizadas pela empresa Olist, uma das líderes no setor de tecnologia e soluções para vendas online, no período de 2016 a 2018. Através dessa análise, buscamos responder a algumas perguntas fundamentais, tais como:

- Quais estados apresentam o maior número de clientes?
- Em quais estados há uma presença mais expressiva de vendedores ativos?
- Quais estados se destacam em termos de volume de compras?
- Quais clientes são os maiores gastadores?
- Qual é a média de gastos por estado?
- Até que ponto o valor do frete impacta nas decisões de compra?

Para abordar essas questões, é essencial analisar os Dataframes disponíveis, os quais contêm informações sobre clientes, vendedores, avaliações, e outros aspectos que abrangem o ciclo completo, desde a compra até a entrega do produto.

Ao examinar esses Dataframes, identificamos variáveis cruciais para compreender e responder às perguntas mencionadas, e até mesmo para formular novas questões. No entanto, é imperativo verificar cada variável, pois podem existir valores ausentes, tipos incorretos ou informações imprecisas.

```python
clientes.join(pedidos,lsuffix='_x', rsuffix='_y').drop('customer_id_y', axis = 1)
```

```python
pedidos_completos = pd.merge(left=clientes,
                             right=pedidos,
                             left_index=True,
                             right_index=True,
                             suffixes=['_x','_y'])\
    .drop('customer_id_y', axis = 1)\
    .rename(columns = {'customer_id_x': "customer_id"})\
    .merge(itens_pedidos,
           left_index=True,
           right_index=True,
           suffixes=['_x','_y'])\
    .drop('order_id_y', axis = 1)\
    .rename(columns={'order_id_x':'order_id'})\
    .merge(pagamentos,
           left_index=True,
           right_index=True,
           suffixes=['_x','_y'])\
    .drop('order_id_y', axis = 1)\
    .rename(columns={'order_id_x':'order_id'})\
    .merge(produtos,
           left_index=True,
           right_index=True,
           suffixes=['_x','_y'])\
    .drop("product_id_y", axis = 1)\
    .rename(columns={"product_id_x":"product_id"})\
    .merge(vendedores,
           left_index=True,
           right_index=True,
           suffixes=['_x','_y'])\
    .drop('seller_id_y', axis=1)\
    .rename(columns={'seller_id_x':'seller_id'})

pedidos_completos
```

# Limpeza de dados

```python
def analise_primaria(df):
    colunas = [i for i in df.columns]
    totais =  [df[i].unique().shape[0] for i in df.columns]
    diferenca = [df.shape[0] - i for i in totais]
    vazios = [df[i].isnull().sum() for i in df.columns]
    df_unicos = pd.DataFrame({"colunas":colunas,"valores_unicos":totais, "diferença": diferenca, "valores_vazios": vazios})
    display(df_unicos)
```

Os dados estão divididos em DataFrames, sendo que cada DataFrame traz algum tipo de informação. Portanto, é necessário examinar um por um para compreender melhor como as vendas estão distribuídas neste período.

```python
clientes.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (14,5))
ax = sns.countplot(data=clientes.drop_duplicates(subset="cliente_id_unico"), x = "cliente_estado", order=clientes['cliente_estado'].value_counts()[:10].index,palette="mako")
sns.despine(left= True)
plt.title("Os 10 estados com maior quantidade de clientes")
plt.xlabel("Estádos",fontsize = 14)
plt.ylabel("Quantidade", fontsize=14)
plt.yticks([])
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., 
            height+400,
            "{:1.0f}".format((height)),
            ha="center")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(14,5))

ax = sns.countplot(data=clientes.drop_duplicates(subset="cliente_id_unico"), x = "cliente_cidade", order=clientes['cliente_cidade'].value_counts()[:10].index,palette="mako")
sns.despine(left=True)
plt.title("As 10 cidades com a maior quantidade de clientes.")
plt.xlabel("Cidades", fontsize = 14)
plt.ylabel("Quantidade", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks([])
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., 
            height+150,
            "{:1.0f}".format((height)),
            ha="center")
plt.show()
```
